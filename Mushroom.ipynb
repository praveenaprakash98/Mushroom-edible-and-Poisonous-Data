 ###  CORRELATION HEATMAP 

```python
# plot Feature Correlation Matrix
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    return np.sqrt(chi2 / (n * (min(confusion_matrix.shape) - 1)))

# Calculate correlations
features = df.columns.drop('class')
correlation_matrix = pd.DataFrame(index=features, columns=features)

for i, feature1 in enumerate(features):
    for j, feature2 in enumerate(features):
        if i <= j:
            correlation_matrix.loc[feature1, feature2] = cramers_v(df[feature1], df[feature2])
            correlation_matrix.loc[feature2, feature1] = correlation_matrix.loc[feature1, feature2]

correlation_matrix = correlation_matrix.astype(float)

plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', square=True, cbar_kws={'shrink': 0.8})
plt.title('Feature Correlation Matrix (Cramér\'s V)', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

    * population and habitat have high correlations with multiple features
	* There are several feature pairs with correlations above 0.5, indicating moderate to strong relationships
    * Many mushroom characteristics are interconnected, which makes biological sense (e.g., habitat might influence spore characteristics)
	* The relatively high correlations suggest some features might be redundant for classification purposes
	* Features like odor, gill-spacing, and cap-surface show varying correlation patterns

### FEATURE IMPORTANCE 

```python
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder

# Encode all categorical variables
df_encoded = df.copy()
encoders = {}
for col in df_encoded.columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    encoders[col] = le

X = df_encoded.drop('class', axis=1)
y = df_encoded['class']

chi2_scores, p_values = chi2(X, y)
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Chi2_Score': chi2_scores,
    'P_Value': p_values
}).sort_values('Chi2_Score', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(data=feature_importance.head(15), x='Chi2_Score', y='Feature', palette='viridis')
plt.title('Top 15 Most Important Features (Chi-square scores)', fontsize=14, fontweight='bold')
plt.xlabel('Chi-square Score')
plt.tight_layout()
plt.show()
```

    1. Ring-type (~2000 score) - The type of ring around the stem is the strongest predictor
	2. Gill-size (~1500 score) - The size of the mushroom's gills is highly significant
	3. Bruises (~1000 score) - Whether the mushroom bruises when handled is important
    
	* There's a clear hierarchy of feature importance, with ring-type being dramatically more important than other features
	* Physical characteristics (ring-type, gill-size, bruises) dominate the top features
	* Color-related features (stalk-color variations, odor) have relatively low importance
	* The gap between the top 3 features and the rest suggests these are the critical identifying characteristics

### TOP FEATURES vs TARGET ANALYSIS

```python
top_features = feature_importance.head(8)['Feature'].tolist()

fig, axes = plt.subplots(2, 4, figsize=(20, 12))
axes = axes.ravel()

for i, feature in enumerate(top_features):
    ct = pd.crosstab(df[feature], df['class'], normalize='index') * 100
    ct.plot(kind='bar', ax=axes[i], color=['lightcoral', 'lightgreen'], 
            title=f'{feature.replace("-", " ").title()}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Percentage')
    axes[i].legend(['Poisonous', 'Edible'])
    axes[i].tick_params(axis='x', rotation=45)

plt.suptitle('Distribution of Top 8 Features by Mushroom Class', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
